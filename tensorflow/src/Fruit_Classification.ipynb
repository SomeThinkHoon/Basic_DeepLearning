```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping
#from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras import optimizers

import numpy as np
import matplotlib.pyplot as plt
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                  horizontal_flip = True,
                                  width_shift_range= 0.1,
                                  height_shift_range = 0.1)
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_data
```

```python
train_generator = train_datagen.flow_from_dataframe(
    dataframe=train_data,
    x_col = 'file',
    y_col = 'label',
    class_mode='categorical',
    target_size=(100,100),
    batch_size=50,
    seed = 13
)
```

```python
test_generator = test_datagen.flow_from_dataframe(
    dataframe=val_data,
    x_col = 'file',
    y_col = 'label',
    class_mode='categorical',
    target_size=(100,100),
    batch_size=50,
    seed = 13,
    shuffle=False
)
```

```python
train_generator = train_datagen.flow_from_directory(
    './data/fruit_Classification/train/train/',
    target_size = (100, 100),  # 위에서 shape이 100,100이라서 100, 100으로 지정
    batch_size=50,   # 알츠하이머 데이터가 각 폴더에 80개 정도 있을 때, 배치 사이즈를 5로 했으니 그 5배로 지정
    class_mode='categorical')  # 'sparse_categorical_crossentropy' 처럼 멀티-레이블 클래스인데, 레이블 인코딩된 형태에 사용
```

```python
test_generator = test_datagen.flow_from_directory(
    './data/fruit_Classification/test/',
    target_size = (100, 100),  # 위에서 shape이 100,100이라서 100, 100으로 지정
    batch_size=25,   # 알츠하이머 데이터가 각 폴더에 80개 정도 있을 때, 배치 사이즈를 5로 했으니 그 5배로 지정
    class_mode='categorical')  # 멀티-레이블 클래스인데, 원-핫 인코딩된 형태에 사용
```

```python
## 일단 CNN으로!!

model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=(100, 100, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))
model.summary()
```

```python
model.compile(loss = 'categorical_crossentropy',
              optimizer = optimizers.Adam(learning_rate = 0.0002),
              metrics = ['accuracy'])
```

```python
## 일단 학습 ㄱㄱ

history = model.fit(
    train_generator,
    epochs = 10,
    validation_data = test_generator,
    validation_steps = 10
)
```

```python
## 학습 성과를 그래프로 그리기

import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize = (12, 4))

sns.lineplot(x = range(len(history.history['loss'])),
             y = history.history['loss'], ax = axes[0],
             label = "Training Loss")

sns.lineplot(x = range(len(history.history['loss'])),
             y = history.history['val_loss'], ax = axes[0],
             label = "Validation Loss")

sns.lineplot(x = range(len(history.history['accuracy'])),
             y = history.history['accuracy'], ax = axes[1],
             label = "Training accuracy")

sns.lineplot(x = range(len(history.history['accuracy'])),
             y = history.history['val_accuracy'], ax = axes[1],
             label = "Validation accuracy")

axes[0].set_title("Loss")
axes[1].set_title("Accuracy")

sns.despine()
plt.show()
```
