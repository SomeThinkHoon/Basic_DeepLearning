```python
import numpy as np
```

```python
raw_data = np.genfromtxt('./data/x09.txt', skip_header=36)
raw_data
```

```python
## 생김새를 3차원으로 확인해보자!!

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline

xs = np.array(raw_data[:, 2], dtype=np.float32)
ys = np.array(raw_data[:, 3], dtype=np.float32)
zs = np.array(raw_data[:, 4], dtype=np.float32)

fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs)
ax.set_xlabel('Weight')
ax.set_ylabel('Age')
ax.set_zlabel('Blood fat')
ax.view_init(15, 15)
plt.show()
```

```python
# 학습 대상의 데이터를 추리자

x_data = np.array(raw_data[:, 2:4], dtype=np.float32)
y_data = np.array(raw_data[:, 4], dtype=np.float32)

y_data = y_data.reshape((25, 1))
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(1, input_shape=(2, ))
])

model.compile(optimizer='rmsprop', loss='mse')
```

```python
model.summary()

## 찾야아할 파라메타는 x1, x2, b 총 3개이다.
## 입력층은  2개이며, 출력층은 1개이다.
```

```python
## fit 학습!!

hist = model.fit(x_data, y_data, epochs = 5000)
```

```python
plt.plot(hist.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.show()
```

```python
plt.plot(hist.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.xlim(4000, 5000)
plt.show()
```

```python
model.predict(np.array([100,44]).reshape(1, 2))

# 첫번째 데이터에 몸무게(100)를 넣어보고 두번째에 나이(44)를 넣어본 결과이다.
```

```python
model.predict(np.array([60,25]).reshape(1, 2))
```

```python
W_,  b_ = model.get_weights()
print('Weight is :', W_)
print('bias is :', b_)
```

```python
## 모델이 잘 만들어졌는지 확인하기 ㅜ이해 데이터를 만들어보자!!

x = np.linspace(20, 100, 50).reshape(50,1)
y = np.linspace(10, 70, 50).reshape(50,1)

X = np.concatenate((x, y), axis =1)
Z = np.matmul(X, W_) +b_
```

```python
## 그리기 시도!!

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs)
ax.scatter(x, y, Z)
ax.set_xlabel('Weight')
ax.set_ylabel('Age')
ax.set_zlabel('Blood fat')
ax.view_init(15, 15)
plt.show()
```
